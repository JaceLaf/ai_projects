## Search


A general search pseudocode is as follows:

```
Here's a useful pseudocode for search, in general:

    Start with a frontier that contains the initial state
    Start with an empty explored set
    Repeat:
        If the frontier is empty, then no solution
        Remove a node from the frontier
        If node contains goal state, return the solution
        Add the node to the explored set
        Expand node, add resulting nodes to the frontier if they aren't already in the frontier or the explored set
```

Now, this is the general idea. How we deal with the frontier (implementation, queue, priority queue, stack, list, dictionary, and so on) and some of the conditions is going to differentiate whether this leads to breadth-first, depth-first, or A$^*$.


In class we talked about alphabetical order, but here we're dealing with locations in a list of lists. Let moves dictate the order. To arrive at the expected output, do not amend the order in which actions are arranged in moves (Up, Right, Down, Left).

````python
def get_neighbors(world:list[list[str]], cell:tuple, moves:list) -> list[tuple]:
    '''
    Finds the accessible next cells and returns a list of their coordinates.

    Args:
        world (list[list[str]]): A 2D grid representing the search area where:
            - Empty cells are represented by ‚¨ú'
            - Obstacles are represented by '‚¨õ'
            - Other characters may be used for different cell types
        cell (tuple): A pair of the coordinates for the given cell
        moves (list): A list of possible moves
    
    Returns:
        neighbors (list[tuple]): A list of accessible coordinates
    '''
    rows = len(world)
    cols = len(world[0]) if rows > 0 else 0
    r, c = cell

    neighbors = []

    # Iterates through possible moves
    for dr, dc in moves:
        nr, nc = r + dr, c + dc

        # Bounds check
        if 0 <= nr < rows and 0 <= nc < cols:
            # Obstacle check
            if world[nr][nc] != '‚¨õ':
                neighbors.append((nr, nc))
    return neighbors


def mark_final_path(world:list[list[str]], parent:dict, start:tuple, goal:tuple) -> None:
    '''
    Reconstructs path based on parent dictionary and marks cells in world with 'üü©'

    Args:
        world (list[list[str]]): A 2D grid representing the search area where:
            - Empty cells are represented by ‚¨ú'
            - Obstacles are represented by '‚¨õ'
            - Other characters may be used for different cell types
        parent (dict): A dictionary containing the previous visited cell for each cell explored
        start (tuple): A pair of the coordinates for the starting cell
        goal (tuple): A pair of the coordinates for the end cell
    '''

    current = goal

    while current is not None:
        r, c = current

        # Skips start and goal
        if current != start and current != goal:
            world[r][c] = 'üü©'

        current = parent.get(current)

    return None

````

````python
def breadth_first_search(world:list[list[str]], start:tuple, goal:tuple, moves:list) -> list[list[str]]: # AUTOGRADER: DO NOT MODIFY FUNCTION HEADER
    """
    Performs a breadth-first search on a 2D grid world to find a path from start to goal.
                if world[neighbor[0]][neighbor[1]] == '‚¨ú':
                world[neighbor[0]][neighbor[1]] = 'üü®'
    Visualizes the search process by marking explored cells with 'üü®' and the final path
    with 'üü©' in the world grid. The function prints the world state after the search
    is complete.

    Args:
        world (list[list[str]]): A 2D grid representing the search area where:
            - Empty cells are represented by ‚¨ú'
            - Obstacles are represented by '‚¨õ'
            - Other characters may be used for different cell types
        start (tuple): The starting coordinates (row, col) in the grid.
        goal (tuple): The goal coordinates (row, col) in the grid.
        moves (list): A list of possible moves

    Returns:
        world (list[list[str]]): The updated world grid with:
        - Cells marked with 'üü®' appeared in the frontier
        - Final path marked with 'üü©'
        - Original world elements (obstacles, robot, goal) remain unchanged
    """
    frontier = []
    explored = set()
    parent = {}

    frontier.append(start)
    explored.add(start)
    parent[start] = None

    while len(frontier) != 0:
        current = frontier.pop(0)

        # Found goal
        if(current == goal):
            break

        # Iterates through neighbors and explores
        for neighbor in get_neighbors(world, current, moves):
            if neighbor not in explored:
                explored.add(neighbor)
                parent[neighbor] = current
                frontier.append(neighbor)

            if world[neighbor[0]][neighbor[1]] == '‚¨ú':
                world[neighbor[0]][neighbor[1]] = 'üü®'
    
    # Marks final path
    if goal in parent:
        mark_final_path(world, parent, start, goal)

    return world
````

## Adversarial Search


The MINIMAX pseudocode is as follows:

```
MINIMAX(state) returns action
    return argmax action in Actions MIN-VALUE(RESULT(state,action))
    

MAX-VALUE(state) returns value
if TERMINAL(state) return UTILITY(state)
v <- -inf
for each action in Actions(state)
    v <- MAX(v, MIN-VALUE(RESULT(state, action))
return v


MIN-VALUE(state) returns value
if TERMINAL(state) return UTILITY(state)
v <- +inf
for each action in Actions(state)
    v <- MIN(v, MAX-VALUE(RESULT(state, action))
return v
```


## Notes. 

* MINIMAX is used to model the opponent as a rational player that will also try it's best to win. When moves are simulated in the tree, that's just what they are, simulations. The actual board only changes once MINIMAX returns an action and that action is taken.
* You will probably need to maintain a deep copy of the world.
* Use a depth parameter to control the number of times the algorithm ping pongs between the calls to the min_value and max_value functions. A depth of 2 would mean P1_MAX's (caller of MINIMAX) moves are simulated and then P2_MIN's moves are simulated, and that's it.
